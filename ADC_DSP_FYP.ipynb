## Applying FFT

```python
signal_data = data9['CH1']

# Compute the FFT of the signal
fft_values = np.fft.fft(signal_data)
fft_freq = np.fft.fftfreq(len(signal_data), 1/fs)

# Compute the magnitude of the FFT (amplitude spectrum) and limit to positive frequencies
fft_magnitude = np.abs(fft_values)
positive_freqs = fft_freq > 0
fft_magnitude = fft_magnitude[positive_freqs]
fft_freq = fft_freq[positive_freqs]

# Plot the FFT amplitude spectrum
plt.figure(figsize=(10, 5))
plt.plot(fft_freq, fft_magnitude, color='blue')
plt.title('FFT Amplitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## Notch Filter

```python
from scipy import signal

fs = 8000.0  # Sample frequency (Hz)
f0 = 50.0  # Frequency to be removed from signal (Hz)
bw = 20.0
Q = f0/bw  # Quality factor
# Design notch filter
b, a = signal.iirnotch(f0, Q, fs)

# Frequency response
freq, h = signal.freqz(b, a, fs=fs)
# Plot
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
ax[0].plot(freq, 20*np.log10(abs(h)), color='blue')
ax[0].set_title("Frequency Response")
ax[0].set_ylabel("Amplitude (dB)", color='blue')
ax[0].set_xlim([0, 100])
ax[0].set_ylim([-50, 10])
ax[0].grid(True)
ax[1].plot(freq, np.unwrap(np.angle(h))*180/np.pi, color='green')
ax[1].set_ylabel("Angle (degrees)", color='green')
ax[1].set_xlabel("Frequency (Hz)")
ax[1].set_xlim([0, 100])
ax[1].set_yticks([-90, -60, -30, 0, 30, 60, 90])
ax[1].set_ylim([-90, 90])
ax[1].grid(True)
plt.show()
```
